```python
batting_first=df[df['WonBy']!=0]
batting_first.head()
```

```python
#plot a histogram for won by
plt.figure(figsize=(7,7))
plt.hist(batting_first['WonBy'])
plt.title('Winings with bat and run')
plt.show()
```

```python
batting_first['WinningTeam'].value_counts()
```

```python
plt.figure(figsize=(7,7))
x=list(batting_first['WinningTeam'].value_counts()[0:3].keys())
y=list(batting_first['WinningTeam'].value_counts()[0:3])
plt.bar(x,y,color=['red','blue','green'])
plt.show()
```

```python
#making pie plot
plt.figure(figsize=(14,14))
plt.pie(batting_first['WinningTeam'].value_counts(), labels=batting_first['WinningTeam'].value_counts().keys(),autopct='%0.1f%%')
plt.show()
```
